Ex 1: Write a program that uses a while loop (not a for loop) to read through a string and print
the characters of the string one-by-one on separate lines.

```python
 # The string to be processed
text = "Sai Htoo Kyaw"

# Initialize a counter/index variable starting at 0
index = 0

# The loop runs until the index is no longer less than the length of the string
while index < len(text):
    # Print the character at the current index
    print(text[index])

    # Manually increment the index to move to the next character
    index += 1
```

Ex 2: A good program will make sure that the data its users enter is valid. Write a program that
asks the user for a weight and converts it from kilograms to pounds. Whenever the user
enters a weight below 0, the program should tell them that their entry is invalid and then ask
them again to enter a weight. [Hint: Use a while loop, not an if statement].

```python
# Conversion factor: 1 kg = 2.20462 lbs
CONVERSION_FACTOR = 2.20462

# Initial input to start the process
weight_kg = float(input("Enter a weight in kilograms: "))

# The loop continues as long as the weight is negative
while weight_kg < 0:
    print("Invalid entry. Weight cannot be less than 0.")
    # Ask again inside the loop to update the variable
    weight_kg = float(input("Please enter a valid weight in kilograms: "))

# Once the loop exits, we know the weight is valid
weight_lbs = weight_kg * CONVERSION_FACTOR

print(f"{weight_kg} kg is equal to {weight_lbs:.2f} pounds.")
```

Ex 3: Write a program that allows the user to enter any number of test scores. The user indicates
they are done by entering in a negative number. Print how many of the scores are Aâ€™s (90 or
above). Also print out the average

```python
# Initialize counters and accumulators
total_sum = 0
count_all = 0
count_as = 0

# Get the first score
score = float(input("Enter a test score (negative to quit): "))

# Continue looping until a negative number is entered
while score >= 0:
    # Update totals
    total_sum += score
    count_all += 1

    # Check if the score is an A
    if score >= 90:
        count_as += 1

    # Get the next score from the user
    score = int(input("Enter the next score (negative to quit): "))

# Display results only if at least one score was entered
if count_all > 0:
    average = total_sum / count_all
    print("\n--- Results ---")
    print(f"Total number of A's: {count_as}")
    print(f"The average score is: {average:.2f}")
else:
    print("No scores were entered.")
```

Ex 4: Write a program that asks the user for a string and a letter. Using a for loop, the
program should print the index of the first occurrence of that letter and a message if the
string does not contain the letter.

```python
# Get input from the user
text = input("Enter a string: ")
target_letter = input("Enter a letter to search for: ")

# We use a flag to track if we found the letter
found_index = -1

# Loop through the indices of the string
for i in range(len(text)):
    if text[i] == target_letter:
        found_index = i
        break  # Exit the loop immediately after finding the first occurrence

# Check the results
if found_index != -1:
    print(f"The first occurrence of '{target_letter}' is at index {found_index}.")
else:
    print(f"The letter '{target_letter}' was not found in the string.")
```

Ex 5: Write a program in which you have a list that contains seven integers that can be 0 or 1. Find the first zero entry in the list and change it to a 1. If there are no zero entries, print a message saying so.

```python
# A list of seven integers (0s or 1s)
numbers = [1, 1, 0, 1, 0, 1, 1]

# We use a flag to track if we successfully flipped a bit
found_zero = False

# Loop through the list using indices
for i in range(len(numbers)):
    if numbers[i] == 0:
        numbers[i] = 1   # Change the zero to a one
        found_zero = True
        break            # Stop after finding the FIRST zero

# Output the results
if found_zero:
    print("Modified list:", numbers)
else:
    print("There are no zero entries in the list.")
```
