Ex 2: Write a function that takes an integer n and returns a random integer with exactly n digits. For
instance, if n is 3, then 125 and 593 would be valid return values, but 093 would not because
that is really 93, which is a two-digit number.

```python
import random

def random_n_digit(n):
    if n <= 0:
        raise ValueError("n must be a positive integer")

    # The smallest n‑digit number is 10^(n-1)
    # The largest n‑digit number is 10^n - 1
    low = 10**(n - 1)
    high = 10**n - 1

    return random.randint(low, high)
print(random_n_digit(3))
```

Ex 3: Write a function called number_of_factors that takes an integer and returns how many
factors the number has.

```python
import math

def number_of_factors(x: int) -> int | float:
    """
    Return the number of positive factors (divisors) of x.

    - For x == 0, returns float('inf') since every nonzero integer divides 0.
    - For other values, counts positive divisors of |x|.
    """
    if x == 0:
        return float('inf')
    n = abs(x)
    if n == 1:
        return 1

    count = 0
    root = int(math.isqrt(n))
    for d in range(1, root + 1):
        if n % d == 0:
            count += 2  # d and n//d are both divisors
    if root * root == n:
        count -= 1  # perfect square: counted the sqrt twice
    return count
print(number_of_factors(6))
```

Ex4: Write a function called first_diff that is given two strings and returns the first location in
which the strings differ. If the strings are identical, it should return -1.Ignore the case in your function. It means h and H are the same.

```python
def first_diff(a: str, b: str) -> int:
    """
    Return the first index at which strings a and b differ (case-insensitive).
    If they are identical ignoring case, return -1.
    """
    la, lb = a.lower(), b.lower()
    min_len = min(len(la), len(lb))

    for i in range(min_len):
        if la[i] != lb[i]:
            return i

    # If all shared characters match but lengths differ,
    # the first difference is at the end of the shorter string.
    if len(la) != len(lb):
        return min_len

    return -1
```

Ex6: Write a function that solves the system of equations ax + by = e and cx +dy = f. Your function should takes 6 parameters, (a, b, c, d, e,f) and return the values of x and y.

```python
def solve_2x2(a, b, c, d, e, f):
    """
    Solve the linear system:
        a*x + b*y = e
        c*x + d*y = f

    Returns:
        (x, y) as a tuple of floats.

    Raises:
        ValueError: if the system has no unique solution (determinant == 0).
    """
    det = a * d - b * c
    if det == 0:
        raise ValueError("No unique solution: determinant is zero (lines are parallel or coincident).")
    x = (e * d - b * f) / det
    y = (a * f - e * c) / det
    return x, y
```

Ex7:  Write a function called change_case that given a string, returns a string with each upper  case letter replaced by a lower case letter and vice-versa

```python
def change_case(s):
    result = ""
    for char in s:
        if char.islower():
            result += char.upper()
        elif char.isupper():
            result += char.lower()
        else:
            result += char   # keep non‑letters unchanged
    return result
```

Ex8:  Write a function called primes that is given a number n and returns a list of the first n  primes. Let the default value of n be 100.

```python
def primes(n: int = 100) -> list[int]:
    """
    Return a list of the first n prime numbers.
    Default n is 100.

    For n <= 0, returns an empty list.
    """
    if n <= 0:
        return []

    result = []
    candidate = 2

    def is_prime(x: int) -> bool:
        if x < 2:
            return False
        if x % 2 == 0:
            return x == 2
        # Test odd divisors up to sqrt(x)
        d = 3
        while d * d <= x:
            if x % d == 0:
                return False
            d += 2
        return True

    while len(result) < n:
        if is_prime(candidate):
            result.append(candidate)
        candidate += 1

    return result
```

Ex9: A Tic-tac-toe board can be represented be a 3×3 two-dimensional list, where zeroes stand for  empty cells, ones stand for X’s and twos stand for O’s. Write a function that is given such a list and randomly chooses a spot in which to place  a 2 (O). The spot chosen must currently be empty and a spot must be chosen.

```python
import random

def place_random_O(board):
    """
    Given a 3x3 board (list of lists) where:
      - 0 = empty
      - 1 = X
      - 2 = O
    Randomly choose an empty spot (0) and place a 2 (O) there.

    Modifies the board in place and also returns (row, col).

    Raises:
        ValueError: if there are no empty spots.
    """
    # Collect all empty positions
    empties = [(r, c)
               for r in range(3)
               for c in range(3)
               if board[r][c] == 0]

    if not empties:
        raise ValueError("No empty spots available to place O.")

    r, c = random.choice(empties)
    board[r][c] = 2
    return r, c
```

Ex10:  Write a function that is given a 9 × 9 potentially solved Sudoku and returns True if it is solved correctly and False if there is a mistake. The Sudoku is correctly solved if there are
 no repeated numbers in any row or any column or in any of the nine “blocks.”


```python
def valid_sudoku(board):
    """
    Given a 9x9 Sudoku board (list of lists),
    return True if it is correctly solved, otherwise False.
    """

    # Helper: returns True if "group" contains digits 1..9 exactly once
    def valid_group(group):
        return sorted(group) == list(range(1, 10))

    # Check rows
    for row in board:
        if not valid_group(row):
            return False

    # Check columns
    for c in range(9):
        col = [board[r][c] for r in range(9)]
        if not valid_group(col):
            return False

    # Check 3x3 blocks
    for br in range(0, 9, 3):      # block row start (0,3,6)
        for bc in range(0, 9, 3):  # block col start (0,3,6)
            block = [
                board[r][c]
                for r in range(br, br + 3)
                for c in range(bc, bc + 3)
            ]
            if not valid_group(block):
                return False

    return True
```
