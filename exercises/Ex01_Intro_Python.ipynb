# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
print("Enter details for Apple:")
apple_quantity = int(input("Enter quantity of apples: "))
apple_price = float(input("Enter price per apple: "))
apple_total = apple_quantity * apple_price

print("\nEnter details for Orange:")
orange_quantity = int(input("Enter quantity of oranges: "))
orange_price = float(input("Enter price per orange: "))
orange_total = orange_quantity * orange_price

print("\nEnter details for Mango:")
mango_quantity = int(input("Enter quantity of mangoes: "))
mango_price = float(input("Enter price per mango: "))
mango_total = mango_quantity * mango_price

total_cost = apple_total + orange_total + mango_total

print(f"\nTotal cost for all fruits: ${total_cost:.2f}")
```

# Ex 6: Random Number
Write a program that generates a random number, x, between 1 and 50, a random number y between 2 and 5, and computes x^y.

```python
import random
x = random.randint(1, 50)
y = random.randint(2, 5)
result = x ** y
print(f"Random number x: {x}")
print(f"Random number y: {y}")
print(f"{x}^{y} is: {result}")
```

# Ex 8: Computing.
Write a program that asks the user to enter two numbers, x and y, and computes |x-y|/(x+y).

```python
x = float(input("Enter value for x: "))
y = float(input("Enter value for y: "))
if x + y == 0:
    print("Error: The sum of x and y cannot be zero (division by zero).")
else:
    result = abs(x - y) / (x + y)
    print(f"The result of |{x} - {y}| / ({x} + {y}) is: {result:.4f}")
```

# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
total_seconds = int(input("Enter the number of seconds: "))
minutes = total_seconds // 60
remaining_seconds = total_seconds % 60
print(f"{total_seconds} seconds is {minutes} minutes and {remaining_seconds} seconds.")
```

# Ex 10: Using Math Module

Write a program that asks the user for a number and then prints out the sine, cosine, and tangent of that number.

```python
import math
angle_degrees = float(input("Enter an angle in degrees: "))
angle_radians = math.radians(angle_degrees)
sn = math.sin(angle_radians)
cs = math.cos(angle_radians)
tn = math.tan(angle_radians)
print(f"\nResults for {angle_degrees}°:")
print(f"Sine:    {sn:.4f}")
print(f"Cosine:  {cs:.4f}")
print(f"Tangent: {tn:.4f}")
```
