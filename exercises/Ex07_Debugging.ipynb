```python
### Q2) Write a program that generates 10000 random numbers between 1 and 100 [inclusive] and counts how many of them are multiples of 12.
# This program has errors, and the result differs from the expected. Debug the code.
import numpy as np
count = 0
range_start = 1
range_end = 100
total_samples = 10000
for i in range(total_samples):
    num = np.random.randint(range_start, range_end+1)

    if num%12==0:
        count=count+1

print('Number of multiples of 12:', count)

```

```python
#Q3) The following program should counts how many of the squares of the numbers from 1 to 100 end in 1.  (Example: 1,81,121).
# ## Debug the code.
import numpy as np
def fun_square_count():
    num_list = np.arange(1, 101)
    # Square the numbers
    squared_num = [value**2 for value in num_list]
    count = 0

    for num in squared_num:
        # FIX: Changed // (floor division) to % (remainder)
        if (num % 10 == 1):
            count += 1
    return count

result = fun_square_count()
print(f"Your result is {result}")
print(f"Expected result is 20")
```

```python
#Q4) The following program should check if a number is a prime number. Debug the code.
import numpy as np

def is_prime(n):
    if n < 2: return False
    for num in range(2, n):
        if n % num == 0:
            return False
    return True

numbers = np.arange(3, 21)
results = {}
for num in numbers:
    results[num.item()] = is_prime(num)

print(f"Your results: {results}")
```

```python
#Q5) The following code uses stub functions to design the game. Complete the code.

import random

def generate_random_num():
    """
    Generates a random integer between 1 and 100 (inclusive).
    """
    # Use random.randint for a simple inclusive range [1, 100]
    return random.randint(1, 100)

def get_user_input():
    """
    Asks the user to guess the number and ensures it is an integer.
    """
    while True:
        try:
            guess_num = int(input("Guess a number between 1 and 100: "))
            return guess_num
        except ValueError:
            print("That's not a valid integer. Please try again.")

def check_the_number(actual_num, guess_num):
    """
    Checks if the actual number and guess number are the same.
    """
    # Return True if they match, otherwise False
    return actual_num == guess_num

def display_the_feedback(is_correct, chances):
    """
    Prints feedback based on the correctness of the guess.
    """
    if is_correct:
        print("Bingo!")
    else:
        print(f"Wrong guess! You have {chances} chances left.")

### Do not change the functions below ###
def guess_num_game():
    actual_num = generate_random_num()

    max_chances = 10
    for counter in range(1, max_chances+1):
        guess_num = get_user_input()
        is_correct = check_the_number(actual_num, guess_num)
        display_the_feedback(is_correct, max_chances - counter)
        if(is_correct):
            return "You did it! You won the game"

    return "You lost! Play one more time? "

def main():
    while True:
        win_lost_msg = guess_num_game()
        print(win_lost_msg)
        try:
            is_continue = int(input("Enter 1 to continue (any other number to quit): "))
            if is_continue != 1:
                print("Bye. Thanks for playing.")
                break
        except ValueError:
            print("Invalid input. Exiting game.")
            break

if __name__ == "__main__":
    main()
```
